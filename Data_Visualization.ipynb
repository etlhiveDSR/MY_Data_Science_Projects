# Univariate analysis 
    Categorical columns
        1. Countplot
        2. Countplot(horizontal)
        3. Pie chart

```python
import pandas as pd
A = pd.read_csv("/users/bajajvbh/Desktop/Vaibhav/Datasets/Cars93.csv")

A = A.drop(labels=["id","Make","Model"],axis=1)
cat = []
con = []
for i in A.columns:
    if(A[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
A.AirBags.value_counts()
```

```python
A.AirBags.value_counts().plot(kind="bar")
```

```python
A.AirBags.value_counts().plot(kind="barh")
```

```python
A.AirBags.value_counts().plot(kind="pie")
```

```python
nrows = round(len(cat)/3,0)+1
```

```python
cat
```

```python
plt.figure(figsize=(20,20))
w = 1
for i in cat:
    plt.subplot(nrows,3,w)
    A[i].value_counts().plot(kind="pie")
    w = w + 1
```

```python
plt.figure(figsize=(20,20))
w = 1
for i in cat:
    plt.subplot(nrows,3,w)
    A[i].value_counts().plot(kind="barh")
    w = w + 1
```

```python
plt.figure(figsize=(20,20))
w = 1
for i in cat:
    plt.subplot(nrows,3,w)
    A[i].value_counts().plot(kind="bar")
    w = w + 1
```
